```python
from sklearn.model_selection import train_test_split
```

```python
x=data1
```

```python
x= x.drop(columns='Heart Disease')
```

```python
y=x['heart dise']
```

```python
x=x.drop(columns='heart dise')
```

```python
from sklearn.naive_bayes import GaussianNB
naiveBayes=GaussianNB()
```

```python
trainX,testX,trainY,testY=train_test_split(x,y,random_state=1)
```

```python
naiveBayes.fit(trainX,trainY)
```

```python
pred=naiveBayes.predict(testX)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(pred,testY))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec_tree = DecisionTreeClassifier()
```

```python
dec_tree.fit(trainX,trainY)
```

```python
dc_pred=dec_tree.predict(testX)
```

```python
print(accuracy_score(dc_pred,testY))
```
